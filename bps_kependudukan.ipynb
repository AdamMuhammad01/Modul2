```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d
```

```python
a =pd.read_excel('indo.xls', na_values=['-'],header=3, skipfooter=3)
a.rename(columns={'Unnamed: 0':'Provinsi'},inplace=True)
a = a.fillna(0)
tahun=[]
for i in a.columns[::1]:
    tahun.append(i)
a
```

```python
plt.style.use('seaborn')
fig = plt.figure('3d', figsize=(16,16))
p = plt.subplot(111,projection='3d')
plt.suptitle('Penduduk Indonesia Tahun 1971', size=15)

x= np.arange(len(a['Provinsi']))
y= np.arange(len(a['Provinsi']))
z=np.zeros(len(a['Provinsi']))
dx=np.ones(len(a['Provinsi']))
dy=np.ones(len(a['Provinsi']))
dz=a[1971]
p.bar3d(x,y,z,dx,dy,dz, color='pink')
```

```python
import pandas as pd
df = pd.read_csv('pokemon-spawns.csv')
df = df[:100]
df

import folium
map = folium.Map(
      location =[df['lat'][0],df['lng'][0]],
      zoom_start=15,
      tiles = 'openstreetmap'
)
for i in range(len(df)):
    folium.Marker(
    location=[df['lat'][i],df['lng'][i]],
    tooltip=f"<img src=https://raw.githubusercontent.com/PokeAPI/sprites/master/sprites/pokemon/{df['num'][i]}.png></img"
    ).add_to(map)
```

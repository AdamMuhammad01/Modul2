```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib notebook
```

```python
tm = pd.read_csv('TLKM1.JK.CSV', parse_dates = ['Date'],index_col='Date')
isat = pd.read_csv('ISAT.JK.CSV', parse_dates = ['Date'],index_col='Date')
fr = pd.read_csv('fren.JK.CSV', parse_dates = ['Date'],index_col='Date')
xl = pd.read_csv('excl.JK.CSV', parse_dates = ['Date'],index_col='Date')
```

# No.1

```python
plt.figure(figsize=(12,5),dpi=300)
plt.plot(tm.Close, color='red')
plt.plot(isat.Close, color='blue')
plt.plot(fr.Close,color='lightblue')
plt.plot(xl.Close,color='green')
plt.xlabel("Tanggal")
plt.suptitle('Grafik Harga Saham Provider Telekomunikasi di Indonesia',
         fontdict={'family':'calibri', 'size':20, 'color':'black'},y = 1.1)
plt.xticks(rotation=45, size=6)
plt.ylabel("Rupiah IDR")
plt.legend(["PT Telekomunikasi Indonesia Tbk", "PT Indosat tbk", "PT Smartfren Telecom Tbk","PT XL Axiata Tbk"],
           bbox_to_anchor=(0., 1.02, 1., .102), loc='upper left', ncol = 4, mode="expand", borderaxespad=0.)



```

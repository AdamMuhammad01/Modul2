```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from matplotlib import style

plt.figure('SPLTV',figsize=(10,5))
custom=plt.subplot(121,projection='3d')
#x-2y+z=6
x1=np.array([1,-2,1])
y1=np.array([5,3,7])
z1=np.array([6])
custom.scatter(x1,y1,z1)
custom.set_xlabel('X')
custom.set_ylabel('Y')
custom.set_zlabel('Z')
```

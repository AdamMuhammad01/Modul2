```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as im
df = pd.read_csv('Datasales.csv')
```

```python
df  = df.pivot(index = 'nama', columns = 'mobil')
df
```

```python
plt.style.use('seaborn')
fig, p =plt.subplots()
plt.imshow( df, cmap = 'hot')
plt.colorbar()
```

```python
plt.style.use('seaborn')
fig, p =plt.subplots()
plt.imshow( df, cmap = 'hot')
plt.colorbar()
col = list(map(lambda x:x[1], df.columns.tolist()))
print(col)
i = list(map(lambda x: x,df.index.tolist()))
print(i)
```

```python
plt.style.use('classic')
fig, p =plt.subplots()
plt.imshow( df, cmap = 'hot')
plt.colorbar()
col = list(map(lambda x:x[1], df.columns.tolist()))
print(col)
i = list(map(lambda x: x,df.index.tolist()))
print(i)
p.set_xticks(np.arange(len(col)))
p.set_xticklabels(col)
p.set_yticks(np.arange(len(i)))
p.set_yticklabels(i)
plt.show()
```

```python
plt.style.use('classic')
fig, p =plt.subplots()
plt.imshow( df, cmap = 'BuPu_r')  # 'r' = reverse
plt.colorbar()
col = list(map(lambda x:x[1], df.columns.tolist()))
i = list(map(lambda x: x,df.index.tolist()))

# isi nilai tiap kotak

for x in range(len(i)):
    for y in range(len(col)):
        p.text(y, x, df.iloc[x, y],fontsize = 20, ha= 'center', va='center', color='y')
        
p.set_xticks(np.arange(len(col)))
p.set_xticklabels(col)
p.set_yticks(np.arange(len(i)))
p.set_yticklabels(i)
# p.set_ylim(2.5,-0.5) #biar ga kepotong
plt.show()
```

```python
ab = pd.read_csv('Medalatlet.csv')
```

```python
ab
```

```python
dfc = ab.corr()   #mencari korelasi dari data
dfc
```

```python
plt.style.use('seaborn')
plt.imshow(dfc, cmap='magma')
plt.colorbar()

col = list(map(lambda x:x, dfc.columns.tolist()))
plt.xticks(np.arange(len(dfc.index)),
           dfc.index.tolist(),
          rotation=90
          )

plt.yticks(np.arange(len(dfc.index)),
           dfc.index.tolist())

for x in range(len(col)):
    for y in range(len(col)):
        plt.text(y, x, round(dfc.iloc[x, y], 3), ha= 'center', va='center', color='g')
        
plt.show()
```

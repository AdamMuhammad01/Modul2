```python
bcg = pd.read_csv('Balita Terimunisasi BCG 1995-2017.csv',na_values=['n.a'])
bcg.rename(columns={'% Balita yang pernah mendapat imunisasi BCG':'bc'}, inplace=True)
campak = pd.read_csv('Balita Terimunisasi Campak 1995-2017.csv',na_values=['n.a'])
campak.rename(columns={'% Balita yang pernah mendapat imunisasi Campak':'ca'}, inplace=True)
dpt = pd.read_csv('Balita Terimunisasi DPT 1995-2017.csv',na_values=['n.a'])
dpt.rename(columns={'% Balita yang pernah mendapat imunisasi DPT':'dp'}, inplace=True)
polio = pd.read_csv('Balita Terimunisasi Polio 1995-2017.csv',na_values=['n.a'])
polio.rename(columns={'% Balita yang pernah mendapat imunisasi Polio':'po'}, inplace=True)
```

```python
a = bcg.interpolate(method='linear')
b = campak.interpolate(method='linear')
c = dpt.interpolate(method='linear')
d = polio.interpolate(method='linear')
```

```python
plt.figure(figsize=(15,18),dpi=100)

plt.style.use('seaborn')
plt.subplot(2,2,1)
plt.bar(a.Tahun, a.bc, color='r')
plt.xticks(rotation=90)
plt.title('BCG', color='black')
plt.xticks(rotation=0, size=10)
plt.subplot(2,2,2)
plt.bar(b.Tahun, b.ca, color='g')
plt.title('Campak', color='black')
plt.subplot(2,2,3)
plt.bar(c.Tahun, c.dp, color='y')
plt.title('DPT', color='black', size=11) #y = -0.4

plt.subplot(2,2,4)
plt.bar(c.Tahun, d.po, color='b')
plt.title('Polio', color='black', size=11)
```

```python
x = ab.interpolate(method='linear')
y = cd.interpolate(method='linear')
z = ef.interpolate(method='linear')
f = gh.interpolate(method='linear')
```

```python
plt.figure(figsize=(15,18),dpi=100)

plt.style.use('seaborn')
plt.subplot(2,2,1)
plt.bar(bcg.Tahun, x, color='r')
plt.title('BCG', color='black',size=15)
plt.subplot(2,2,2)
plt.bar(campak.Tahun, y, color='g')
plt.title('Campak', color='black',size=15)
plt.subplot(2,2,3)
plt.bar(dpt.Tahun, z, color='y')
plt.title('DPT', color='black',size=15)
plt.subplot(2,2,4)
plt.bar(polio.Tahun, f, color='b')
plt.title('Polio', color='black',size=15)
```

```python
# %matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.subplot(221)
plt.plot([2,2,1], [2,4,3])
plt.title('Tes 1')

plt.subplot(224)
plt.plot([2,4,1], [1,4,3])
plt.title('Tes 2')
plt.suptitle('super title')
plt.show()
```

```python
plt.style.use('classic')
df =pd.DataFrame({
    'x':[1,2,3,4,5,6],
    'y':[2,6,8,5,3,1],
    'z':[1,2,6,8,9,0]
})
# plt.figure('Tes',figsize=(3,3))
plt.plot(df['x'], df[['y','z']])
plt.grid(True)
plt.xticks(np.arange(6), list('ab,cb,cd,de,gf,gh'), rotation=180)
plt.yticks(np.arange(0,10,step=1))
plt.title('Tes Chart')
plt.show
```

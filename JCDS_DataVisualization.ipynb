```python
ase2 = "select name,Population from country where Region = 'southeast asia'  "
ba = pd.read_sql(ase2,alamatDB)
ba
```

```python
Nama = ba['name']
Populasi = ba['Population']
colors =['r','r','g','r','pink','brown','silver','silver','black','green','blue']
```

```python
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(8,5),dpi=100)
plt.pie(Populasi, labels=Nama, colors=colors, autopct='%1.1f%%', shadow=True, startangle=0)
plt.title('Persentase Negara Asean',
         fontdict={'family':'calibri', 'size':15, 'color':'black'})
plt.show()
```

```python
ase3 = "select Name,Population,GNP from country where Region = 'southeast asia' order by Name asc "
bc = pd.read_sql(ase3,alamatDB)
bc
```

```python
Negara = bc['Name']
GNP = bc['GNP']
```

```python
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(8,4),dpi=100)
a = plt.bar(Negara, GNP,color=['b','orange','g','r','purple','brown','pink','silver','black','green','yellow'],
       edgecolor ='black',linewidth=5)
plt.scatter(Negara, GNP,
       edgecolor ='green',linewidth=5)
plt.title('Populasi Negara Asean',
         fontdict={'family':'calibri', 'size':15, 'color':'black'})
# plt.figure(figsize=(10,10))
plt.xlabel('negara')
plt.ylabel('Gross National Product (US$)')
plt.xticks(rotation=90, size=10)
plt.yticks(size=5)
```

```python
ase4 = "select Name,SurfaceArea from country where Region = 'southeast asia' order by Name asc "
bd = pd.read_sql(ase4,alamatDB)
bd
```

```python
Negoro = bd['Name']
Ld = bd['SurfaceArea']
colors =['red','green','black','blue','pink','brown','silver','gold','black','green','blue']
```

```python
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(8,4),dpi=100)
plt.pie(Ld, labels=Negoro, colors=colors, autopct='%1.1f%%', shadow=True, startangle=0)
plt.title('Persentase Luas Daratan Asean',
         fontdict={'family':'calibri', 'size':15, 'color':'Black'})
plt.show()
```
